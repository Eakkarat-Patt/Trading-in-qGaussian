```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def DistributionPlot(func1, func2, title, binwidth):
    plt.figure(figsize=(8, 5), dpi=500)
    sns.histplot(func1, binwidth=binwidth, binrange=[func2.min(), func2.max()], color='r')
    sns.histplot(func2, binwidth=binwidth, binrange=[func2.min(), func2.max()], color='b')
    # plt.xlim([-50, 150])
    plt.xlim([min(func1), max(func1)])
    plt.xlabel('$')
    plt.savefig('C:/Users/Eakka/iCloudDrive\Documents/Research Tsallis statistics/manuscript/Figures/PnL dist.png')
    plt.show()
```

```python
df1 = pd.read_csv('Dataset/profit q=1.011 alpha=0.0001.csv')
df2 = pd.read_csv('Dataset/profit q=1.5 alpha=0.0001.csv')
```

```python
DistributionPlot(df1['mm3'], df2['mm3'],'PnL', 0.2)
```
