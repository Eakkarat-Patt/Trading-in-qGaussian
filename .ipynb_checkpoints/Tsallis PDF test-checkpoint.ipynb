```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.special import gamma
import seaborn as sns
import pandas as pd
```

```python
def x(start, stop, n):
    return np.linspace(start, stop, n)

def c(q):
    return (np.pi * gamma(1 / (q - 1) - 0.5) ** 2) / ((q - 1) * gamma(1 / (q - 1)) ** 2)

def B(q, t):
    return c(q) ** ((1 - q) / (3 - q)) * ((2 - q) * (3 - q) * t) ** (-2 / (3 - q))

def Z(q, t):
    return ((2 - q) * (3 - q) * c(q) * t) ** (1 / (3 - q))

def Pq(q, t, start, stop, n):
    return ((1 - B(q, t) * (1 - q) * x(start, stop, n) ** 2) ** (1 / (1 - q))) / Z(q, t)
```

```python
Z(1.5, 1)
```

```python
Pq(1.4, 1)
```

```python
plt.plot(x(-4, 4, 10000), Pq(1.2, 1, -4, 4, 10000))
plt.plot(x(-4, 4, 10000), Pq(1.4, 1, -4, 4, 10000))
plt.plot(x(-4, 4, 10000), Pq(1.6, 1, -4, 4, 10000))
plt.plot(x(-4, 4, 10000), Pq(1.6, 5, -4, 4, 10000))
```
