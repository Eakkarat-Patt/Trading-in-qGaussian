```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('Data/S&P 500 daily.csv', thousands=',') # Daily price
df['Date'] = pd.to_datetime(df['Date'], format='%b %d, %Y')
df.set_index(['Date'],inplace=True)
df['Return'] = (df['Price']-df['Price'].shift(1))/df['Price'].shift(1)
df['Log return'] = np.log(df['Price']/df['Price'].shift(1))
df.drop(['Open', 'High', 'Low', 'Vol.', 'Change %'], inplace=True, axis=1)
```
