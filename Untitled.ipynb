```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import pandas as pd
```

```python
start = '2000-01-01'
end = None
```

```python
df = pd.read_csv('spx_d.csv')[['Date', 'Close']]
df = df.set_index('Date')
if end == None:
    df = df.loc[start:]
else:
    df = df.loc[start:end]
df['Log return'] = np.log(df['Close']/df['Close'].shift(1))
mu = df['Log return'].mean()
sigma = df['Log return'].std()
df['Log return'] = (df['Log return']-mu)/sigma  # Standardization
normalFit = np.linspace(np.min(df['Log return']), np.max(df['Log return']), 10000)
plt.figure(figsize=(8,5), dpi=400)
plt.plot(normalFit, norm.pdf(normalFit, 0, 0.6))
plt.hist(df['Log return'], density=True, bins=150)
plt.xlim(-2.5, 2.5)
plt.show()
```
