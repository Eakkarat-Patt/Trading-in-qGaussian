```python
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def TimeSeriesPlot(x, y1, y2, pathNum, stop, legend=True, ylabel=None, label1=None, label2=None):
    plt.figure(figsize=(8, 5), dpi=100)
    plt.plot(x[x <= stop], y1[pathNum, x <= stop], label=label1)
    plt.plot(x[x <= stop], y2[pathNum, x <= stop], label=label2)
    plt.xlim([0, stop])
    # plt.ylim([1.06, 1.11])
    plt.ylabel(ylabel)
    plt.xlabel('Time')
    if legend:
        plt.legend()
    plt.show()
    
def TimeSeriesPlot2(x, y1, y2, y3, pathNum, stop, label1=None, label2=None, label3=None, legend=True, ylabel=None):
    plt.figure(figsize=(8, 5), dpi=500)
    plt.plot(x[x <= stop], y1[pathNum, x <= stop], label=label1)
    plt.plot(x[x <= stop], y2[pathNum, x <= stop], label=label2)
    plt.plot(x[x <= stop], y3[pathNum, x <= stop], label=label3)
    plt.xlim([0, stop])
    # plt.ylim([1.06, 1.11])
    plt.ylabel(ylabel)
    plt.xlabel('Time')
    if legend:
        plt.legend()
    plt.show()

def DistributionPlot(func1, func2, title, label1, label2, binwidth):
    plt.figure(figsize=(8, 5), dpi=100)
    sns.histplot(func1[:, -1], binwidth=binwidth, binrange=[min(func1[:, -1].min(), func2[:,-1].min()), max(func1[:, -1].max(), func2[:,-1].max())], color='r', label=label1)
    sns.histplot(func2[:, -1], binwidth=binwidth, binrange=[min(func1[:, -1].min(), func2[:,-1].min()), max(func1[:, -1].max(), func2[:,-1].max())], color='b', label=label2)
    # plt.xlim([-50, 150])
    plt.xlim([min(func1[:, -1]), max(func1[:, -1])])
    plt.title(title)
    plt.legend()
    plt.show()
    
def CashPlot(time,y1, y2, order1, order2, pathNum):
    plt.figure(figsize=(8, 5), dpi=200)
    plt.plot(time, y1[pathNum, :], label='GBM')
    plt.plot(time, y2[pathNum, :], label='qGaussian')
    for i in range(0, time.shape[0]):
        if order1[pathNum, i] == 1:
            a = plt.scatter(time[i], y1[pathNum, i], color='b', s=9)
        if order1[pathNum, i] == 2:
            b = plt.scatter(time[i], y1[pathNum, i], color='g', s=9)
        if order1[pathNum, i] == 3:
            c = plt.scatter(time[i], y1[pathNum, i], color='r', s=9)
        if order1[pathNum, i] == 4:
            d = plt.scatter(time[i], y1[pathNum, i], color='k', s=9)
        if order2[pathNum, i] == 1:
            plt.scatter(time[i], y2[pathNum, i], color='b', s=9)
        if order2[pathNum, i] == 2:
            plt.scatter(time[i], y2[pathNum, i], color='g', s=9)
        if order2[pathNum, i] == 3:
            plt.scatter(time[i], y2[pathNum, i], color='r', s=9)
        if order2[pathNum, i] == 4:
            plt.scatter(time[i], y2[pathNum, i], color='k', s=9)
    plt.xlim([0.0, time[-1]])
    plt.title('Cash vs time')
    plt.ylabel('Cash')
    plt.xlabel('Time')
    plt.legend()
    plt.legend((a, b, c, d), ('buy', 'sell', 'No matched order', 'both order matched'))
    plt.show()
```

```python
cash2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 Cash.txt')
cash3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Cash.txt')
time = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Time.txt')
order2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 Order.txt')
order3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Order.txt')
deltaA2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 DeltaA.txt')
deltaA3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 DeltaA.txt')
deltaB2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 DeltaB.txt')
deltaB3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 DeltaB.txt')
rvPrice2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 Reservation price.txt')
rvPrice3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Reservation price.txt')
probA2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 ProbA.txt')
probA3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 ProbA.txt')
probB2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 ProbB.txt')
probB3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 ProbB.txt')
price2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 Price.txt')
price3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Price.txt')
inventory2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 Inventory.txt')
inventory3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Inventory.txt')
profit2 = np.loadtxt('Data/Meeting 9 8 July 22/mm2 Profit.txt')
profit3 = np.loadtxt('Data/Meeting 9 8 July 22/mm3 Profit.txt')
```
